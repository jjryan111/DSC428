```python
import torch
import matplotlib.pyplot as plt
import numpy

def f(x):
    return x**3 + 3*x**2 - 2*x + 1

```

```python
x = torch.linspace(-10, 10, 50, requires_grad=True)
y = f(x)
y.sum().backward()
dy_dx = x.grad
```

```python
plt.figure()
plt.plot(x.detach().numpy(), y.detach().numpy(), label="f(x)")
plt.plot(x.detach().numpy(), dy_dx.detach().numpy(), label="dx/dy")
plt.legend()
plt.show()
```
